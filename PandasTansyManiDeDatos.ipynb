```python
import pandas as pd
```

```python
datos= pd.read_json('/content/datos_hosting.json')
```

```python
datos.head()
```

```python
datos= pd.json_normalize(datos['info_inmuebles'])
```

```python
datos.head()
```

```python
#Viendo las columas dentro de mi DataFrame
columnas= list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[3:])
datos
```

```python
datos.reset_index(drop=True, inplace=True)
datos
```

##Datos numericos

```python
#Chequeamos para ver el tipo de dataframe y corregir en caso de que sea un tipo que no deberia
datos.info()
```

```python
#Convirtiendo datos objeto a numericos enteros
import numpy as np
datos['max_hospedes']=datos['max_hospedes'].astype(np.int64)
```

```python
#La agrupamos para hacerlo más eficiente
listanumerica =['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos[listanumerica]= datos[listanumerica].astype(np.int64)
```

```python
#Convirtiendo datos objeto a decimales sin caracteres.
datos['evaluacion_general']=datos['evaluacion_general'].astype(np.float64)
```

```python
#Convirtiendo datos objeto a numeros que contienen caracteres, como $
datos['precio']
```

```python
#Usaremos lambda y llamaremos cada elemento de nuestra columna precio como x, de aqui le llamaremos a la funcion replace, lo primero que se encuentre en el parentesis entre comillas simples es lo que queremos quitar
#strip es para borrar espacios en blanco, cuando remplazamos el dolar o la coma por nada pueden dejarnos espacios vacios
datos['precio']=datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip()).astype(np.float64)
```

```python
#Cuando vamos a cambiar varias series (columnas), usamos applymap, ya de una vez pondremos astype para cambiar el tipo de dato
datos[['cuota_deposito', 'cuota_limpieza']]=datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip()).astype(np.float64)
```

##Datos tipo texto

###Tokenizacion

```python
#Vamos a toquenizar, que es sacar pequeñas palabras o frases que nos permiten tener datos.
#Lo primero será seleccionar una columna, como nuestros datos estan como objeto deberemos avisar que es string, usaremos .str
#Usaremos el metodo para strings "lower" para ponerlo todo en minusculas, queremos hacerlo lo más simple posible
datos['descripcion_local']=datos['descripcion_local'].str.lower()
```

```python
datos['descripcion_local']
```

```python
#Todo lo que está dentro de los corchetes es el regex, el primer simbolo (⌃) significa negacion, todo lo contrario será cambiado
#(a-zA-Z0-9) le estaremos diciendo que esto NO lo reemplace
#\- nos servira para que entienda que tambien queremos mantener el guion, si no le ponemos la barra, puede interprestarse que queremos separar
#\' queremos que mantenga la comilla simple.
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r"[^a-zA-Z0-9\-']", ' ', regex=True)
```

```python
#Aqui estamos diciendo que si el guion no tiene palabra o caracter a los lados (izquierda y derecha) lo reemplace por nada.
datos['descripcion_local']=datos['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)',' ', regex=True)
```

```python
#split es un metodo que ya separa cada elemento y lo hace una lista
#entre parentesis ira el separador de cada elemento, por defecto es espacio
datos['descripcion_local']=datos['descripcion_local'].str.split()
```

```python
datos['comodidades']=datos['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
datos['comodidades']=datos['comodidades'].str.split(',')
```
